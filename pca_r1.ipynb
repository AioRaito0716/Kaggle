```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train_data.csv',nrows=100000)

dfl = pd.read_csv('train_labels.csv',nrows=100000)
```

```python
df.head(30)
```

```python
df_col = df.columns
```

```python
ID_list = list(df['customer_ID'].unique())
```

```python
dfl_ex = dfl.iloc[:len(ID_list),:]
```

```python
dfl_col = dfl.columns
```

```python
len(ID_list)
```

```python
label_list = list(dfl_ex['target'])
```

```python
label_dict = {}
for ID,label in zip(ID_list,label_list):
    label_dict[ID] = label
```

```python
grouped = df.groupby('customer_ID')
```

```python
avg_df = pd.DataFrame(grouped.mean())
```

```python
acols = avg_df.columns
med_avg = {}
for col in acols:
    med_avg[col] = avg_df[col].median()
    avg_df[col].fillna(med_avg[col],inplace=True)

avg_df.dropna(how='all',axis=1,inplace=True)
```

```python
var_df = pd.DataFrame(grouped.var())
med_var = {}
for col in acols:
    med_var[col] = var_df[col].median()
    var_df[col].fillna(med_var[col],inplace=True)
    
var_df.dropna(how='all',axis=1,inplace=True)
```

```python
vcm_df = avg_df.corr()
vcm_df.dropna(how='all',inplace=True)
vcm_df.dropna(how='all',axis=1,inplace=True)
```

```python
vcv_df = var_df.corr()
vcv_df.dropna(how='all',inplace=True)
vcv_df.dropna(how='all',axis=1,inplace=True)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
X = np.array(vcm_df)
pca.fit(X)
```

```python
feature = pca.fit_transform(avg_df)
```

```python
l_ratio = pd.DataFrame(pca.explained_variance_ratio_,index=['PC{}'.format(i+1) for i in range(len(vcm_df))])
```

```python
r_list = list(l_ratio[0])
sum_ratio = 0
sum_list = []
for rate in r_list:
    sum_ratio += rate
    sum_list.append(round(sum_ratio,3))
```

```python
#第10主成分までで87%、20までで92%、30までで95%
print(sum_list[10],sum_list[20],sum_list[30])
```

```python
pca_v = PCA()
Xv = np.array(vcv_df)
pca_v.fit(Xv)

feature_v = pca_v.fit_transform(vcv_df)
v_ratio = pd.DataFrame(pca_v.explained_variance_ratio_,index=['PC{}'.format(i+1) for i in range(len(vcv_df))])

r_list_v = list(v_ratio[0])
sum_ratio_v = 0
sum_list_v = []
for rate in r_list_v:
    sum_ratio_v += rate
    sum_list_v.append(round(sum_ratio_v,3))
```

```python
#第10主成分までで87%、20までで92%、30までで95%
print(sum_list_v[10],sum_list_v[20],sum_list_v[50])
```

```python
#第20主成分までの主成分得点
#第20主成分までの固有ベクトル×各IDの説明変数（avg_dfの各列）
avg_df.drop('D_87',axis=1,inplace=True)
```

```python
a_index = avg_df.index
```

```python
np_avg = np.array(avg_df)
```

```python
feature = pca.fit_transform(np_avg,)
```

```python
#var_df.drop('D_87',axis=1,inplace=True)
np_var = np.array(var_df)
feature_v = pca_v.fit_transform(np_var)
```

```python
#特徴量：平均のみ
X = feature[:,:30]
Y = label_list
```
